```python
import getpass as gp
from pyspark.sql import SparkSession
```

```python
user = gp.getuser()
spark = SparkSession.builder \
            .appName(f'{user}-week4-assignment-question-3') \
            .master('yarn') \
            .getOrCreate()
```

```python
spark
```

## Create Base RDD

```python
rdd_review_file_input = spark.sparkContext.textFile('/public/trendytech/reviews/trendytech-student-reviews.csv') \
                            .flatMap(lambda row : row.split(' '))
```

```python
rdd_review_file_input.take(10)
```

```python
local_boring_words = set(line.strip().lower() for line in open('/data/trendytech/boringwords.txt', 'r'))
```

```python
local_boring_words
```

```python
boring_words_broadcast = spark.sparkContext.broadcast(local_boring_words)
```

```python
boring_words_broadcast.value
```

```python
rdd_students_review = rdd_review_file_input \
                        .filter(lambda words : words.lower() not in boring_words_broadcast.value) \
                        .map(lambda words : (words.lower(),1)) \
                        .reduceByKey(lambda x,y : x+y) \
                        .sortBy(lambda col : col[1], ascending= False)
```

```python
rdd_students_review.take(20)
```
